# ‡∏î‡∏∂‡∏á1‡∏´‡∏ô‡πâ‡∏≤

```python
page = 4
data = requests.get('https://www.livinginsider.com/searchword/Condo/Buysell/'+str(page)+'/%E0%B8%A3%E0%B8%A7%E0%B8%A1%E0%B8%9B%E0%B8%A3%E0%B8%B0%E0%B8%81%E0%B8%B2%E0%B8%A8-%E0%B8%82%E0%B8%B2%E0%B8%A2-%E0%B8%84%E0%B8%AD%E0%B8%99%E0%B9%82%E0%B8%94.html')
soup = bs4.BeautifulSoup(data.text)
count = 0
post_list =[]
price_list =[]
location_list =[]
nearbts_list = []
nearuni_list = []
size_list =[]
floor_list =[]
bed_list =[]
bath_list =[]
for c in soup.find_all('div',{'class':'col-md-3 col-sm-4 col-xs-6'}):
  try:
    count += 1
    post = c.find('div',{'class':'col-md-12 col-sm-12'})    
    post_list.append(post.find('a').text.split('\n')[2].strip())

    price_list.append(int(c.find('div',{'class':'t-16'}).text.split('‡∏ø')[1].replace(',','')))
    size_list.append(float(c.find('div',{'class':'col-sm-offset-1 col-md-offset-1 ic-detail col-md-5 col-sm-5 col-xs-6 '}).text.split()[0]))
    floor_list.append(c.find('div',{'class':' col-md-5 col-sm-5 col-xs-6 ic-detail '}).text.split()[1])
    bed_list.append(c.find('div',{'class':'col-xs-6 ic-detail col-md-5 col-sm-5 col-sm-offset-1 col-md-offset-1 '}).text.split('\n')[2].strip())
    
    if  c.find('div',{'class':'col-xs-6 ic-detail col-md-5 col-sm-5 '}).text.split()[0] == "‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤":
      bath_list.append(int(c.find('div',{'class':'col-xs-6 ic-detail col-md-5 col-sm-5 '}).text.split()[1]))
    else:
      bath_list.append(int(c.find('div',{'class':'col-xs-6 ic-detail col-md-5 col-sm-5 '}).text.split()[0]))

    location_temp = c.find('div',{'class':'col-md-12 col-sm-12 col-xs-12 ic-detail-zone'}).text.split()
    location = ""
    for k in location_temp:
      location += str(k) + " "
    #remove front last space   
    location_list.append(location.strip())

    nearbts = "no"
    nearuni = "no"
    facilities = c.find_all('img')
    facilities
    for a in facilities:
      try:        
        if a['title'] == 'Near BTS/MRT':
          nearbts = 'yes'
        if a['title'] == 'Near University':
          nearuni = 'yes'         
      except:
        pass
    nearbts_list.append(nearbts)
    nearuni_list.append(nearuni)    
  except:
    print("block data value: " + str(count)  )
print(post_list)
print(price_list)
print(size_list)
print(floor_list)
print(bed_list)
print(bath_list)
print(location_list)
print(nearbts_list)
print(nearuni_list)
```

```python
table = pd.DataFrame([post_list,price_list,size_list,floor_list,bed_list,bath_list,location_list,nearbts_list,nearuni_list]).transpose()
table.columns = ['post','price','size','floor','room type','no. bathroom','location','near bts/mrt','near univercity']
table
```
